## Lists & Strings

```python
# a collection of data : List
purchases = [] # empty
# append an entry
purchases.append(99) # add entry to back of list
purchases.append(34) # add entry to back of list
purchases.append(22) # add entry to back of list
print(purchases)
# In Python collections (e.g. a list) are zero-indexed => first element is at index 0
# insert an entry at front of queue
purchases.insert(0, 55)
print(purchases)
# extend a list
purchases.extend([5,6,7,8,9,10,11])
purchases.insert(-1, 999)
print(purchases, sum(purchases), sum(purchases) / len(purchases), max(purchases), min(purchases))

mid = len(purchases) // 2
print(purchases[mid])

start=0
stop_before=2
print(purchases[:stop_before])
```

### Slicing

```python
# slicing [start:stop:step]
L1 = purchases[:3] # start at the beginning i.e. index 0
L2 = purchases[3:] # continue to the end of the list i.e. index len(list)-1
L3 = purchases[:] # start at beginning continue to the end (i.e. make a copy of the list)
L4 = purchases[::3] # give me every third elemement of list
print(L1, L2, L3, L4)
print(purchases)
# slice = list[start:stop:step]  list[i]  list[:i] list[i:] list[:] list[::i]
```

## Dictionaries & Sets

### Key Value Pairs & lookup

```python
# dictionary is a key:value pair
eng2ger = dict() # dictionary constructor
print(eng2ger, type(eng2ger))

eng2ger = {
    "key": "value", # key:value pair
    "car": "auto",
    "bicycle": "fahrrad",
    "motorbike": "motorrad",
    "horse": "pferd"
}
print(eng2ger.get("key")) # returns a value because the key:value pair exists
print(eng2ger.get("key2")) # returns None (because it does not exist)

# dictionary has items, keys, and values
print(eng2ger.keys())
print(eng2ger.values())

print('key2' in eng2ger.keys()) # is 'key2' in the key list?

for k, v in eng2ger.items():
    print(f"Key:{k} Value:{v}")

x = 'wayne','karen' # married
print(x)
```

```python
# handling missing entries
key = input("Enter a English: ")
result = eng2ger.get(key)
if result: 
    print(f"{result} is {key} in German")
else:
    print(f"Key {key} not found")
```

### Parameters and Arguments

In programming, parameters and arguments are closely related but serve distinct purposes in the context of functions or methods. Understanding their differences is essential for writing and discussing code effectively.

**Parameters**
Parameters are placeholders defined in the function or method declaration. They specify the input that the function expects to receive when it is called. Parameters act as local variables within the function and are used to process the data passed during execution.

Below, x and y are the parameters of the add function. They define the structure of the input the function can accept.

```python
def add(x, y): # x and y are parameters
   return x + y
```

**Arguments**
Arguments are the actual values or variables passed to the function when it is invoked. These values are assigned to the corresponding parameters during the function call, enabling the function to execute with specific data.

In the example below: case, 5 and 3 are the arguments provided to the add function, which are assigned to the parameters x and y.

```python
result = add(5, 3) # 5 and 3 are arguments
print(result) # Output: 8
```

**Key Differences**
* Definition: Parameters are variables listed in the function declaration, while arguments are the actual values passed during the function call.
* Location: Parameters are part of the function definition, whereas arguments are part of the function invocation.
* Role: Parameters define the input structure, while arguments supply the actual data for processing.
* Naming: Parameters are named identifiers used within the function, while arguments can be literals, variables, or expressions.

While parameters and arguments are often used interchangeably in casual discussions, their distinction lies in their roles within a function. Parameters define the expected input, and arguments provide the actual data during execution. Understanding this difference ensures clarity when writing or explaining code.

### Three Parameter Types

```python
def f_fixedpos(a, b):
    print(a,b)
    return None

def f_variable(*args): # sent as a tuple
    print(args, type(args))
    return None

def f_keyword(**args): # sent as key:value pair = dictionary
    print(args, type(args))
    return None

y = f_fixedpos(10, 20)
y = f_variable(4,3,4,5,6,7,7,45,56,4,3,45,0,56)
y = f_keyword(name="Ben", age=24, module="cybersecurity")
```
