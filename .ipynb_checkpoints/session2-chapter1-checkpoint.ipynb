### Numeric Types 
There are three distinct numeric types: integers, floating point numbers, and complex numbers. In addition, Booleans are a subtype of integers. Integers have unlimited precision. Floating point precision and internal rep

```python
# integers (whole numbers)
a = 5
print(f"a is of type {type(a)}")

# floating point numbers (fractional numbers)
a = 5.0
print(f"a is of type {type(a)}")

# a special note about division
a, b = 11, 2
r1, r2 = a / b, a // b # what's the difference?
print(f"{a} / {b} = {r1} <--- float division")
print(f"{a} // {b} = {r2} <--- floor (integer) division")

# Boolean truth - True or False (a subclass of integer)
wrote_exam = True
passed_exam = False 
print(type(wrote_exam), type(passed_exam)) # True and False are both Booleans

print(f"True is equivalent to {int(True)}") # True is equivalent to integer value 1
print(f"False is equivalent to {int(False)}") # False is equivalent to integer value 0

# Boolean conditions - making comparisons yields the truth
a, b = 10, 6
c1, c2, c3 = a < b, a > b, a == b
print(f"{a} < {b} is {c1}")
print(f"{a} > {b} is {c2}")
print(f"{a} is equal to {b} is {c3}")

```

 

For more detailed information about making comparisons, refer to https://docs.python.org/3/library/stdtypes.html?highlight=comparison 

### Sequence Types
Sequences are *finite*, *ordered collections* of objects *indexed* by non-negative numbers.  They are distinguished by their mutability (i.e. their ability to be changed after they are created).  Some cannot change (aka *immutable* sequences), such as strings, bytes, tuples and ranges.  Some can change (aka *mutable* sequences), such as lists and byte arrays.

1. Finite - they have a measurable length (e.g. how many numbers are there in a list?)
2. Ordered sets - objects can occur earlier or later in a sequence (e.g. first element, last element, nth element)
3. Indexable - individual objects in a sequence can be distinguished by a non-negative index (e.g. item $i$ of sequence $a$ is selected by $a[i]$ )

The sequences we are interested in are either $immutable$ (i.e. unchangable once created; such as strings, tuples and ranges) or $mutable$ (i.e. designed to be changed onced created; such as lists).  The goal of this section is to show how sequences are important to solving problems.  The solutions to these problems will be expressed in Python 3 code.



```python
# Strings - text data
msg = "England expects that every man will do his duty." # 21 October 1805 (Adm Horatio Nelson - Battle of Trafalgar)
print(msg)
msg_bytes = msg.encode()  # string encoded to bytes
print(list(msg_bytes))
msg_text = msg_bytes.decode() # bytes decoded into string
print(msg_text)

# CREATING A SEQUENCE

a = []            # an empty sequence a
a = [1, 2, 3]     # a non-empty sequence a
print(type(a), a) # this is in fact a list

b = ()            # an empty sequence b
b = (1, 2, 3)     # a non-empty sequence b
print(type(b), b) # which is a tuple behind the scenes

c = ""            # an empty sequence d
c = "1, 2, 3, 4"  # a non-empty sequence d
print(type(c), c) # this is also known as a string

# a range object is an efficient sequence which yields values only when needed.
print()
d = range(1,4) # a builtin function is used to create a range object
print(type(d), d)
print(list(d), tuple(d), str(d))

# out of interest.... all of these sequences can be created using builtin functions
a = list()
b = tuple()
c = str()
print(a, b, c)
```

### Set Type
Sets are unordered, finite collections of unique, immutable objects. As such, they cannot be indexed by any subscript. Common uses of sets include membership testing, removing duplicates from a sequence, and computing mathematical operations such as intersection, union, difference, and symmetric difference. Sets are mutable, but frozen sets are not.

```python
# CREATING SETS
A = set()    # empty set
B = {4, 5}   # non-empty set

L = [3,5,7,9]
C = set(L)   # non-empty set with iterable constructor; note the difference in member order

A.add(8) # add a member to a set (i.e. sets are mutable)
A.add(5)

C.remove(5)  # remove a member from a set

print(A, type(A), len(A), 5 in A)
print(B, type(B), len(B), 5 in B)
print(C, type(C), len(C), 5 in C)

# syntax: list = [expression for item in iterable]
# - iterable is a range, list, tuple, or any kind of sequence object
# - item is a variable name which takes each value in the iterable
# - expression is a Python expression which is calculated for each value of item

# 1. Create a list of numbers from 1 to 10
L = [x for x in range(1,11)] # create a list consisting of x where x is in the range 1 .. 10
print(L)

# 2. Create a list of squares from 1 to 10
L = [x*x for x in range(1,11)] # create a list consisting of x squared where x is in 1 .. 10
print(L)

# 3. Create a periodic list of values from 1 to 10
L = [x%5 for x in range(15)] # create of list consisting of x modulo 5 where x is in 0..14
print(L)
```

```python
# UNION
A = {3, 4, 5}
B = {4, 5, 6}
C = A | B
print(C)     # <=== {3, 4, 5, 6}

C = A.union(B)
print(C)     # same result

# INTERSECTION
C = A & B
print(C)     # <=== {4, 5}

C = A.intersection(B)
print(C)     # same result


# DIFFERENCE
C = A - B
print(C) # <=== {3} i.e. what is in A that is not in B

C = A.difference(B)
print(C)

C = B - A
print(C) # <=== {6} i.e. what is in B that is not in A

C = B.difference(A)
print(C)

# SYMMETRIC DIFFERENCE
C = A ^ B
print(C) # <=== {3, 6}

C = A.symmetric_difference(B)
print(C) # same result


# DISJOINT SETS

result = A.isdisjoint(B)
print(result)             # <=== False because there are common elements in A and B

A = set([1, 2, 3])
B = set([4, 5, 6])
print(A.isdisjoint(B))    # <=== True because there are NO common elements in A and B
```

### Mapping Types
A mapping object maps hashable values to arbitrary objects. Mappings are mutable objects. There is currently only one standard mapping type, the dictionary. Dictionaries are finite sets of key-value pairs of objects indexed by nearly arbitrary values (i.e. only keys that are hashable).

```python
# DEMO - English-Xhosa dictionary
english_words = ["river", "tree", "mountain", "fire", "boy", "girl"]
xhosa_words = ["umlambo", "umthi", "intaba", "umlilo", "umfana", "intombazana"]

# create a dictionary from these words
e2x = dict(zip(english_words, xhosa_words))
x2e = dict(zip(xhosa_words, english_words))

# display the dictionaries
print(e2x)
print(x2e)

# lookup a Xhosa word for mountain
answer = e2x["mountain"]
print(answer)

```

### SUPER POWER # 1 - COMPREHENSIONS

Comprehensions in Python provide a concise and efficient way to create new sequences or collections from existing ones. They enhance code readability and reduce the need for lengthy loops. Python supports three basic types of comprehensions - viz. list, dictionary and set comprehensions:

```python
# 1. List Comprehensions
squares = [x**2 for x in range(1, 6)]
print(squares) # Output: [1, 4, 9, 16, 25]

evens = [x for x in range(10) if x % 2 == 0]
print(evens) # Output: [0, 2, 4, 6, 8]

# 2. Dictionary Comprehensions
cubes = {x: x**3 for x in range(1, 6)}
print(cubes) # Output: {1: 1, 2: 8, 3: 27, 4: 64, 5: 125}

provinces = ["Gauteng", "KwaZulu-Natal", "Western Cape", "Eastern Cape", "Limpopo", "Mpumalanga", "North West", "Northern Cape", "Free State"]
populations = [15931824, 12312712, 7562588, 7176230, 6402594, 5057662, 4155303, 3044050, 1372943]
total_population = sum(populations)
provincial_populations = {prov:(pop, pop/total_population) for prov, pop in zip(provinces, populations)}
print(provincial_populations)
print("\n2024 Mid-Year Population Estimates")
for province, (pop, pop_ratio) in provincial_populations.items():
    print(f"Province:{province:15s} Population:{pop:11,d} Percentage:{pop_ratio*100:6.1f}%")


# 3. Set Comprehensions
nums = [1, 2, 2, 3, 4, 4, 5, 6]
unique_evens = {x for x in nums if x % 2 == 0}
print(unique_evens) # Output: {2, 4, 6}

squares_set = {x**2 for x in range(1, 6)}
print(squares_set) # Output: {1, 4, 9, 16, 25}
```

### SUPER POWER # 2 - SLICING SEQUENCES
It is often necessary to cut out part of a sequence in the same way a surgeon may do on the operating table.  In the latter case, we call this surgery.  In the former case, we call this slicing.  This is a very common and useful operation that sequences offer.

A slice is a subset of a sequence.  Each slice has:
1. A start - the index of the first value to be included in the slice (starts from zero)
2. An end - the index of the first value to be excluded from the slice (yeah... wierd, I know)
3. An increment - an increment of $n$ will include every $nth$ value in the slice



```python
# basic example
myList = [x for x in range(10, 101, 3)] # every third number from 10 to 100 inclusive
print(myList, len(myList))

# complex example
first_5 = myList[:5]
print("First 5 -", first_5, len(first_5))
last_5 = myList[-5:]
print("Last 5  -", last_5, len(last_5))
mid_point = len(myList)//2
start = mid_point - 2
stop = mid_point + 2
mid_5 = myList[start:stop+1] # +1 because its pythonic
print("Middle 5 -", mid_5, len(mid_5))
```
