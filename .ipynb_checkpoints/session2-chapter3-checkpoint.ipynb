# INTRODUCTION TO PROGRAMMING USING PYTHON 3

In this chapter we examine how to write user-defined functions.

* The idea of a function was developed by Rene Descartes in his book Geometry (1637), used to describe many mathematical relationships. 
* The term "function" was introduced by Gottfried Wilhelm Leibniz almost fifty years later. 
* The idea of a function was further formalized by Leonhard Euler who introduced the notation for a function
  
 


```python
# defining a function in Python 3
def f(x):
    """ Echo function. """
    return x
```

```python
a = 10
b = f(a) # callling the function f with argument a returns the value of b
print(a, b)
```

```python
def g(x):
    """ Increment function. """
    return x+1
```

```python
a = 10
b = g(a) # function g(a) has different effect than f(a)
print(a, b)
```

```python
a = 10
print(a) # a = 10
a = g(a) # inside g() ... value of a is passed to g() and is referred to as 'x' -> g() returns x+1 back to a making it equal 11
print(a)
```

### UPDATED DEMO
Remember the English-to-Xhosa dictionary problem?  Let's refactor that to make use of functions.

```python
# FUNCTION DEMO - word lookup dictionary

def create_word_dictionary(word_list1, word_list2):
    """" Create a dictionary for word-meaning lookups """
    return dict(zip(word_list1, word_list2))

def lookup(dictionary, word):
    """ Use a dictionary to lookup the meaning of a word """
    return dictionary[word]

english_words = ["river", "tree", "mountain", "fire", "boy", "girl"]
xhosa_words = ["umlambo", "umthi", "intaba", "umlilo", "umfana", "intombazana"]

# create a dictionary from these words
e2x = create_word_dictionary(english_words, xhosa_words)
x2e = create_word_dictionary(xhosa_words, english_words)

# display the dictionaries
print(e2x)
print(x2e)

# lookup a Xhosa word for mountain
answer = lookup(e2x, "mountain")
print(answer)

# advanced update ... how do we our functions error free
```

```python
from time import perf_counter, sleep

def slumber(n, verbose=True):
    """ Puts the computer to sleep for n seconds. """
    start = perf_counter()
    for t in range(n):
        if verbose:
            msg = f"Resuming in {n-t:d} \r"
            print(f"{msg}\r", end='')
        sleep(1)
    elapsed = perf_counter() - start
    if verbose:
        msg = f"Elapsed: {elapsed:.1f}s"
        print(f"{msg:30s}")

# call the function
slumber(10)

print("Sleeping for 3 seconds...")
slumber(3, False)
print("Awake again")
```

```python
def f(x):
    print(x, type(x))

def g(*args):
    print(args, type(args))

def h(**kwargs):
    print(kwargs, type(kwargs))
```

```python
print("F - fixed positional arguments")
f(10)
f(5.6)
f("abcd")
print("G - variable arguments")
g(2,4,"6",8,10)
g(1,3,5,'a','b','c',-1, -2,-3)
print("H - keyword arguments")
h(id=10, name="wayne", title="AI Officer", tel=12345675)
```
