### The Needle in a Haystack Problem
You are given a very large firewall log and must find any malicious activity that was successful.
Let's start by first generating a fictional firewall log file...

```python
import random
from time import perf_counter

log_size = 150000  # total number of entries in our fake firewall log file

# Parameters for our fake firewall log
actions = ["ALLOW", "ALLOW", "ALLOW", "ALLOW", "DENY", "BLOCK"]
common_ips = [f"10.0.0.{i}" for i in range(1, 254)]
endpoints = ["/index.html", "/images/logo.png", "/css/style.css", "/about", "/contact"]

# Store the fake database of logs in a list
firewall_logs = []

# generate the log entries
start = perf_counter()
for i in range(log_size):
    time = f"12:{random.randint(10,59)}:{random.randint(10,59)}"
    ip = random.choice(common_ips)
    endpoint = random.choice(endpoints)
    action = random.choice(actions)
    firewall_logs.append(f"{time} - SRC:{ip} -> DST:/api/v1/internal - URI:{endpoint} - STATUS:{action}")

# Inject the ONE malicious line (The Needle)
firewall_logs[random.randint(1500, 3500)] = "12:34:11 - SRC:198.51.100.42 -> DST:/admin/config - URI:/shady_backdoor_exploit.py - STATUS:SUCCESS"
elapsed = perf_counter() - start

print(f"✅ [SUCCESS] Generated {len(firewall_logs):,} active server logs in {elapsed:.2f} seconds.")
```

### Display a small portion of the firewall log

```python
# Let's print just a small sample of the logs to see what the log data looks like

sample_size = 400  # how many entries to display in the sample output

print(f"The first {sample_size} entries are below [remaining {len(firewall_logs)-sample_size:,} lines hidden for your sanity]\n")
print("\n".join(firewall_logs[:sample_size]))
```

### Finding the needle in the cyber haystack

```python
# Let's put this cyber superpower to work to quickly and correctly find the problem

print(f"🔍 Scanning {len(firewall_logs):,} logs for successful exploit signatures...")

# start the clock
start = perf_counter()

# The 3-line superpower
for log in firewall_logs:
    if "SUCCESS" in log:
        print(f"\n🚨 CRITICAL ALERT TRIGGERED:\n{log}")

# stop the clock and reveal how much time this took
elapsed = perf_counter() - start
print(f"\nElapsed: {elapsed:.2f} seconds")
```

## Worked Example: Calculating A Bank Loan Repayment

The task at hand is to calculate how much you must pay each month to your bank, for an amount of money the have loaned to you.

You want to loan R7,200 from the bank.  You want to pay the loan back in 48 months.  The bank will charge you 6% interest annually.  How much money will you pay the bank monthly?

Your goal is to write a script that interacts with the user - getting the necessary inputs to be able to make this calculation, performing the loan repayment calculation, and then presenting the answer to the question to the user.

To aid you in this, I have included the following information about how the loan payment amount is calculated:

$$
PMT = \frac {PV.i.(1+i)^n} {(1+i)^n - 1}
$$

Lets take a look at a script that performs this calculation:

```python
# calculate the monthly repayment amount for a bank loan

# get user input
loan_amount = input('Enter loan amount: ') # this is a floating point value
rate = input('Enter annual interest rate (%): ')  # this is an integer which is divided by 100 and then by 12
term = input('Number of months to repay the loan: ') 

# perform the calculation
PV = float(loan_amount)    # convert loan amount to a float
i = int(rate) / 100 / 12   # convert interest rate to annualized float
n = int(term)              # convert loan duration to integer
t = (1+i) ** n # warning: if the interest rate is zero, then t = 1; this leads to divide by zero in the next line
PMT = (PV * i * t) / (t - 1)

# give the answer to the user
print(f"Loan repayment: R{PMT:.2f} per month.")
```

### Explicit Type Conversions
The script makes use of type conversions to turn the string returned by the input( ) function into a value that can be used in a maths equation.
* *PV = float(loan_amount)* - loan_amount is a string that is returned by *loan_amount = input("Enter loan amount: ")*.  But we need this as a floating point number because the loan_amount variable represents a currency value.  The *float( )* function converts a string into a floating point value.
* *i = int(rate) / 100 / 12* - rate is a string that is returned by *rate = input("Enter annual interest rate (%): ")*.  But we need this as a floating point number to make it useful to this calculation.  The *int( )* function converts a string to an integer (term of loan is in whole months), which is then divded by 100 and by 12 which automatically converts it (implicitly) to a floating point number.

### String Formatting
The output of script *print(f"Loan repayment: R{PMT:.2f} per month.")* provides the answer the user is looking for.  I prefer the "f-string" method of formatting strings.  For an in-depth coverage of how to format string output, review the relevant videos at the end of this notebook.  The user will find more value from a script if it provides the answer in an understandable format - in this case, with an 'R' in front and with the floating point value formatted to 2 decimal places, which is the convention with currency values.

### Simplification
You should always look for ways to simplify your scripts, enhancing readability of your script without sacrificing speed and accuracy.  In this case, I noticed that the equation to calculate the monthly repayment amount had a common term in the numerator and denominator; viz., $(1+i)^n$ so I calculated this term separately, saving the answer in a separate variable *t* and then reused this variable in the final equation.

### Divide by Zero warning
If an interest rate of zero (i=0) is used, then calculating *PMT* will cause a divide-by-zero exception (fatal error which causes your code to crash) because you are dividing by zero: $ (1+i)^n-1 $ will equal zero.
