# WRITING ROBUST CODE

As you get more proficient at writing Python scripts, you will start to encounter errors that either prevent your code from running, or that cause your code to produce unexpected or incorrect results.  Even causing your code to crash suddenly, which is not good.

It is like owning a 4x4 vehicle; as become more confident driving offroad, you will also run the risk of getting stuck in more inaccessible places.


## ERROR TYPES
Initially, you will encounter three main types of errors:
* **Syntax** - when your code violates the rules of the programming language.  A good integrated development environment (IDE) will catch these errors for you.  This is the most basic error.
* **Semantic** - when your code has a logic error.  This causes your code to yield incorrect or unexpected results.  A human programmer must detect and correct these errors.  These errors can and will occur after all syntax errors have been fixed.
* **Run-time** - when your code encounters a scenario which was not anticipated by the programmer during the code development phase.  The software passed inspection with all syntax and semantic errors sorted out, but perhaps the testing of the software was inadequate and as a result the software crashes when it encounters something unanticipated.  Proper software testing is used to ensure that the software is robust (i.e., handles these scenarios without crashing).

## IMPLEMENTING EXCEPTION HANDLING

```python
print(5/4)
print(5/2)
print(5/0)
```

We only have 1 block of code in this trivial example. And the last line of that code has a division by zero error.
To catch this exception, we use the **try** keyword and then indent the misbehaving block of code to make it part of this **try** clause.
We must also write some new code that will run if there is an exception in the block of code linked to the try clause.

```python
try:
    print(5/4)
    print(5/2)
    print(5/0)
except:
    print('Oops.... something went wrong')
```

```python
try:
    print(5/4)
    print(5/2)
    print(5/1) # corrected the code!!!!
except:
    print('Oops.... something went wrong')
else:
    print('No exceptions detected')
```

```python
def test1():
    print('test 1')
    print(5/4)
    print(5/2)
    print(5/1) # corrected the code!!!!

def test2():
    print('test 2')
    print(5/4)
    print(5/2)
    print(5/0) # misbehaving code not corrected!!!!

# we wrap the calls to test1( ) and test2( ) in the try block - this means any error in these functions will be handled by the associated except clause
# note that if an exception occurs, the code in the except block is executed
# note that if no exception occurs, then the code in the else block is executed
# regardless of what happens, the code in the finally block is executed

try:
    test1()
    test2()
except:
    print('Oops.... something went wrong')
else:
    print('No exceptions detected')
finally:
    print('Done')
```

It is not common to use the *else* and *finally* blocks in scripts; but, they are very powerful when they are used.  It should be apparent that our code had fatal run-time errors in it, yet did not crash! This is how you make your code robust.  Whilst this syntax may appear straightforward (or at least it will do with more practice), the process of making your code robust takes practice and experience.  However, you have the tools now to deal with this as you develop more experience throught practice.

Wonderful, our script ran and should have crashed, yet our robust code ensured that the script ran successfully, even though we did not get the good results.  Except (pun intended), the user is not exactly sure what happened in the "oops" scenarios.  What can be done about that?

```python
def calc(x,y):
    """ a more robust function """
    try:
        z = x / y
        print(f"{x} / {y} = {z}")
    except ZeroDivisionError as zde:
        print(f'Oops {zde}')
        return None
    except:
        print('Oops... something went wrong')
        return None
    return z

result1 = calc(5,4)
result2 = calc(5,0)
result3 = calc("5","3")

print(result1, result2, result3)
```
