## Question 10
Give a typical Cartesians system:


Write a Python 3 script that returns the quadrant (i.e., I, II, III or IV) based on the (x,y) cooridinate.

```python
# question 10 - Cartesian Quadrant
coordinates = [(3,5), (7,-3), (-5,7), (-2, -5)]
for coordinate in coordinates:
    x,y = coordinate
    if x == 0 or x == y: 
        quadrant = None
    elif x > 0:
        if y > 0:
            quadrant = "I"
        else:
            quadrant = "IV"
    else:
        if y > 0:
            quadrant = "II"
        else:
            quadrant = "III"
    print(coordinate, quadrant)
```

## Question 2
There is a need for a function that will accept a mathematical operation (provided as a string) and return its actual result (as a number).
For example "2 + 4" should return an integer 6.  Your task is to write 2 functions:
* string_to_number: a single string parameter which must be converted to an integer or float value
* value_of: a single string parameter (a simple mathematical expression) which must be converted to its numerical result (once the operation is performed)


```python
# Question 2 - string functions

def string_to_number(string_value):
    """convert a string into an integer or a float"""
    if string_value.isdigit():
        return int(string_value)
    elif "." in string_value:
        if string_value.replace('.','',1).isdigit():
            return float(string_value)
    return None

def value_of(expression):
    """convert string expression into a number"""
    value = 0
    tokens = expression.split(" ")
    operand1 = string_to_number(tokens[0])
    operand2 = string_to_number(tokens[2])
    operator = tokens[1]
    if operator == "+":
        value = operand1 + operand2
    elif operator == "-":
        value = operand1 - operand2
    elif operator == "*":
        value = operand1 * operand2
    elif operator == "/":
        try:
            value = operand1 / operand2
        except Exception as e:
            print(f"ERROR - {e}")
            value = None
    else:
        print(f"{operator} not recognized")
        value = None
    return value


expressions = ["8 / 4", "9 * 3", "5.5 + 1.5", "9.4 - 3.6", "2.7 / 0.0"]
for expr in expressions:
    print(expr, '=', value_of(expr))
```

## Question 3
Write a script that functions as a very strong password generator.  A very strong password must have a length of 16 characters, and the characters must include lowercase & upper case letters, numbers and punctation marks.  Write a function that returns random passwords that meet this requirement, by generating a password of a specified length (i.e., 16) consisting of a random selection from different character sets (i.e., lower case letters, upper case letters, numbers and punctuation marks).

```python
# question 3 - generator of very strong passwords

import string
import random

lowercase = string.ascii_lowercase
uppercase = string.ascii_uppercase
digits = string.digits
punctuation = string.punctuation
alphabet = lowercase + digits + uppercase + punctuation

def random_choice(characters):
    """return a random character"""
    return random.choice(characters)

def password(length=16):
    """returns a very strong password"""
    pwd = ""
    for _ in range(length):
        pwd += random_choice(alphabet)
    return pwd


pwd = password()
print(pwd)
```

## Question 4
User input validation is an important and common task.  Write a Python 3 script which validate user input - i.e., validates user input and checks if the type of this input is valid; e.g., "2.5" is a float, "22" is an integer, etc.

**Note** - a float can be thought about as an integer with a decimal point inside: "3.456" is "3456" with a decimal point somewhere inside.  Put another way, if my string has a decimal point, and I remove it, if what I am left with is an integer, then what I had is a float.

```python
# question 4 - numeric input validation (using string operations)

def valid_integer(str_int):
    """return True if str_int is an integer otherwise return False"""
    return str_int.isdigit()

def valid_float(str_float):
    """return True if str_float is a float otherwise return False"""
    return valid_integer(str_float.replace('.','',1))

test_subjects = ["35", "7.9483", "2233", "1a.43", "FF", "A1", "3982", "44.32"]
for subject in test_subjects:
    if valid_integer(subject):
        print(subject, "is an integer")
    elif valid_float(subject):
        print(subject, "is a float")
    else:
        print(subject, "is invalid")
```

## Question 5
Write a Python 3 function that accepts a single string parameter (word) and returns the middle character of the word. If the word's length is odd, return the middle character. If the word's length is even, return the middle 2 characters.  For example:
* if s = "boat" then return "oa"
* if s = "boating" then return "t"
* if s = "X" return "X"

```python
# question 4 - middle char(s) extraction

def middle(s):
    """return the middle character(s) from s"""
    if len(s) <= 1: return s
    mid = len(s) // 2
    if len(s) % 2 == 1:
        return s[mid:mid+1]
    return s[mid-1:mid+1]

S = [".", "XOX", "boatmoon", "bloated"]
for s in S:
    print(f"'{s}' => '{middle(s)}'")
```

## Question 6
You have Google stock data in a CSV file ("google_stock_data.csv").  Write a Python 3 function that reads this stock data and returns the stock movement (adjusted close - open price) and the volume of trade for a given date.  Any errors handling the file must be handled so that the script does not crash.  If not data for the given date is found, return None, otherwise return the movement and volume.

```python
# question 6 - csv stock price analysis
import csv

def stock_data(filename, date):
    """return stock movement and volume of trade for given date or return None"""
    results = []
    try:
        with open(filename, 'rt') as f:
            data = csv.reader(f)
            for row in data:
                if row[0] == date: results.append(row)
    except FileNotFoundError:
        print(f"{filename} not found")
    return results
 
dates = ["8/19/2004", "8/19/2009", "8/19/2014", "8/19/2019"]
filename = "google_stock_data.csv"
for dt in dates:
    print(stock_data(filename, dt))

filename = "microsoft_stock_data.csv"
for dt in dates:
    print(stock_data(filename, dt))
```
