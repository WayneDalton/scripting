### B.1 An Interpreted Scripting Language



#### Compiled versus Interpreted Languages
Programming languages typically fall into loosely-defined camps: 
1. Assembly Languages. A low-level language which uses instructions that can be directly executed by the CPU. Each assembly instruction causes the CPU to perform a very specific task (e.g. move data to a register, load data from memory, add two values etc.).  While it is possible to write programs directly in machine code, it is tedious and error prone therefore programs are very rarely written directly in machine code.  Instead assembly, compiled or interpreted languages are used.
2. Compiled Languages. A high-level language that uses a compiler to produce machine code that runs natively on a specific machine architecture (aka "compiled" language).  C is the best-known example of a "compiled" language.
3. Interpreted Languages. A high-level language that uses an interpreter to interpret instructions one at a time and then runs them natively on a specific machine architecture (aka "interpreted" languages.  Python and Java are "interpreted" languages.  CPython's interpreter is written in C. The use of virtual machines in the process of interpretation permits distinguishing between interpreted and emulated languages.

Compiled languages are optimized for speed whilst interpreted languages are optimized for ease of implementation and flexibility (e.g. platform independence, dynamic typing, enhanced readability, etc).  If you have a conversation with someone who speaks your primary/home language "natively", then the conversation flows as fast as the speakers want it to; a higher degree of brevity in speech is possible and frequent use of language features such as metaphor, euphemism and colloquialisms or slang. Contrast having the same conversation with someone who does not speak your language at all.  You would speak through a translator and the dialog would understandably take more time and possibly also be limited in some aspects, but permits a "wider" conversation.

In reality today, the lines between the compiled and interpreted "camps" are rather quite blurred.  For example, C# is considered to be a "compiled" language, but in fact uses both compilers and interpreters as do many other languages.  To quote Wikipedia, "In principle, programs in many languages may be compiled or interpreted, emulated or executed natively, so this designation is applied solely based on common implementation practice, rather than representing an essential property of a language".

#### Interpreting Byte Codes
Similar to Java, Python code is first translated into byte code, which is then interpreted by a CPython virtual machine and executed as machine code that is native to the local machine architecture.  When you code in Python, you will use an editor to produce a text file with a file extension ".py". The following exerise will illustrate what byte code looks like:


1. Open a command line window. This will vary depending on your operating system and distribution of Python.  The standard environment in use during this module is Windows 7 and Anaconda 3; in this case, you will open the Anaconda Prompt.





2. Open Notepad - write a line of Python code. This is the simplest line of code, so we can have something to compile.  The classic "first program" is known as Hello World.  In Python, this is a single line of code; viz. print("Hello, World!")





3. Save your code. You can use Notepad's File menu and select the Save option.  Simply provide a file name (dont forget the .py extension).  In this example, the file name is "hello.py"





4. Compile your code manually. In the screenshot below, you can see that hello.py is in the directory.  Check that your file was saved properly by requesting a directory listing by issuing the command 'dir'.  If the file exists in the local directory, you can then compile it.




The command to compile the code in hello.py actually calls Python.exe and uses an external module (py_compile) to convert the text in hello.py and creates a byte code version of this program.  Python byte code is saved into a special cache directory (__pycache__) in the local subdirectory.

Notice, the byte code file is saved with a ".pyc" file extension.  It also has a postfix to the file name ".cpython-37" which tells us the underlying python reference implementation and version of Python (in this case version 3.7).  CPython is the original byte code interpreter used by the version of Python that ships off the python.org website. 

For the techie geeks, you can get details on how to download and compile your own copy of the interpreter or learn how the interpreter is organized:
1. https://devguide.python.org/setup/


```python
def AddOne(x):
    """ very simple function that returns your input incremented by 1 """
    return x + 1

y = AddOne(5)
print(y) # returns 6
```

To see the byte code for this simple function, we would need this code:

```python
byte_code = AddOne.__code__.co_code # raw bytecode
codes = list(byte_code) # put them into a list for clarity

print(byte_code)
print(codes)
```

To understand what the byte code means, we can dissassemble it:

```python
import dis

# user-friendly display of byte code
print(dis.dis(AddOne))
print('end\n') # so we can separate this output from the one below

# mapping of opcode to opname
for index, code in enumerate(codes):
    print(index, code, '=', dis.opname[code])
```

### B.2 Object-oriented By Design
Python is an object-oriented programming language; all data in a Python program is represented by objects or by relations between objects.  Every object has:
1. An identity.  This never changes once it has been created; you may think of it as the object’s address in memory.
2. A type. This determines the operations that the object supports and the possible values for objects of that type.
3. A value. Objects whose value can change are said to be mutable; objects whose value is unchangeable once they are created are called immutable.  This is determined by the type of the object; e.g. numbers, strings and tuples are immutable, while dictionaries and lists are mutable.



```python
# whole numbers are called integers; these are objects in Python
# therefore, we can examine this object's identity, type and value

print('identity:', id(6)) # identity of an object x is id(x)
print('type:', type(6)) # type of an object x is type(x)
print('value:', 6) # value of an object x is x
```

```python
# real numbers (fractions) are called floats; these are objects in Python
# therefore, we can examine this object's identity, type and value

print('identity:', id(6.0))
print('type:', type(6.0))
print('value:', 6.0)
```

```python
# text data are called strings; these are objects in Python
# therefore, we can examine this object's identity, type and value

print('identity:', id('hello, world!'))
print('type:', type('hello, world!'))
print('value:', 'hello, world!')
```

Python has support for imaginary numbers and for logical (true or false) types which are also objects.
Typically, when we code we use the type and value of an object and replace an object's identity with a label (a.k.a. a variable name).  We cover topic in more detail in a later section, and consider the side-effects of this design decision.

```python
x = 6
print(id(x), type(x), x)

x = 6.0
print(id(x), type(x), x)

x = "hello, world!"
print(id(x), type(x), x)
```

### B3. Dynamically Typed
You may have noticed in the code above, that I used the same label (x) for three different objects (an integer, a float and a string).  This is rather convenient and part of the reason why Python is a simpler language to learn.  However, this is typically not possible in other languages.




```python
# an object is bound to a name at run-time
x = 1
print(id(x), type(x), x)
```

In the code above, an integer object is created and assigned the name 'x'
The object resides in memory at the memory location returned by id(x)
The object is of a type integer which has the value of 1



```python
# it is possible to bind a name to objects of a different type during program execution (i.e. at run-time)
# here the (variable) name is bound to a floating point object; 1 and 3.14 are different objects with different ids
x = 3.14
print(id(x), type(x), x)
```

Note that the integer object has been decoupled from the name 'x', which has now been assigned to the floating point object.  At some point, the decoupled object will be cleared up by the memory garbage collector process and this memory is then freed up.



The reason the code above works is because Python makes use of dynamic typing.  For more detail refer to https://pythonconquerstheuniverse.wordpress.com/2009/10/03/static-vs-dynamic-typing-of-programming-languages/

Okay, so lets put this new-found knowledge and insight to the test.  Consider the following Python code; run it and think about the output.  Then answer the questions below.

```python
x = 2
print(id(x), type(x), x)

y = x
print(id(y), type(y), y)

y = y + 1
print(id(y), type(y), y)
```

1. Why does variable y share the same identity with variable x in line 2 of the output?
2. Why does the identity of variable y change in line 3 of the output?

It should be clear by now that in the first instance, the label y is assigned to the same object to which x was assigned.  This means that the labels (names) x and y are both pointing to the same memory location.  However, when we update the value of y (y = y + 1) we have created a new object which has its own identity, type and value; which can be clearly seen by the third line of output.



Note: the identity of an object will be dependent on the machine you run it on, so you will get different values for id() when you run this on a different machine.  This is normal and does not alter the underlying principles discussed in this unit.

### B.4 Strongly Typed
A strongly-typed language requires EXPLICIT CONVERSIONS between unrelated types, commonly referred to as type conversion and type coercion.  A type conversion in Python is explicit (i.e. you intentionally change the type) and sometime it happens in the implementation of an operation (which is type coercion).


```python
# both are integers - this will work
a = 10
b = 5
print(type(a), type(b))
print(a+b)

# both are strings - this will work
a = '10'
b = '5'
print(type(a), type(b))
print(a+b)

# mixed types that are unrelated to the operation - this will not work
a = 10
b = '5'
print(type(a), type(b))
print(a+b) # <= ERROR - coercion happens with aligned types only
```

```python
# this works because of the explicit conversion of b to an integer
a = 10
b = '5'
print(type(a), type(b))
print(a + int(b)) # explicit type conversion

```

```python
# interesting case of operator behaviour for related types
# in this case adding an integer and a floating point yields a floating point
a = 10
b = 5.0
print(a+b) # type coersion - handled by the underlying interpreter
print(type(a), type(b), type(a+b))
```
