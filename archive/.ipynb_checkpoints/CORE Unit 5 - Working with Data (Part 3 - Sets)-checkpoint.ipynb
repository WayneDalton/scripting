# Working with Data using Python Data Types
In this unit the focus will be on the data types available as part of the core language.  These are considered as standard types and are built into the interpreter. https://docs.python.org/3/library/stdtypes.html

1. Numeric Types - There are three distinct numeric types: integers, floating point numbers, and complex numbers. In addition, Booleans are a subtype of integers. Integers have unlimited precision. Floating point precision and internal representation for the machine on which your program is running is available in sys.float_info. Complex numbers have a real and imaginary part, which are each a floating point number.


2. Sequence Types - Sequences are finite, ordered collections of objects indexed by non-negative numbers.  They are distinguished by their mutability (i.e. their ability to be changed after they are created).  Some cannot change (aka immutable sequences), such as strings, bytes, tuples and ranges.  Some can change (aka mutable sequences), such as lists and byte arrays.


3. Set Types - Sets are unordered, finite collections of unique, immutable objects. As such, they cannot be indexed by any subscript. Common uses of sets include membership testing, removing duplicates from a sequence, and computing mathematical operations such as intersection, union, difference, and symmetric difference. Sets are mutable, but frozen sets are not.


4. Mapping Types - A mapping object maps hashable values to arbitrary objects. Mappings are mutable objects. There is currently only one standard mapping type, the dictionary. Dictionaries are finite sets of key-value pairs of objects indexed by nearly arbitrary values (i.e. only keys that are hashable).

## A. Set-based Problems
A set object is an unordered collection of distinct hashable objects. Common uses include membership testing, removing duplicates from a sequence, and computing mathematical operations such as intersection, union, difference, and symmetric difference.

### Getting the hash of it
An object is hashable if it has a hash value which never changes during its lifetime, and can be compared to other objects. Hashable objects which compare equal must have the same hash value. Hashability makes an object usable as a dictionary key and a set member, because these data structures use the hash value internally. All of Python’s immutable built-in objects are hashable.

### Sets come in two flavors
There are currently two built-in set types, set and frozenset. The set type is mutable — the contents can be changed using methods like add() and remove(). Since it is mutable, it has no hash value and cannot be used as either a dictionary key or as an element of another set. The frozenset type is immutable and hashable — its contents cannot be altered after it is created; it can therefore be used as a dictionary key or as an element of another set.

Further reading:
* https://docs.python.org/3.7/library/stdtypes.html#set-types-set-frozenset
* https://docs.python.org/3.7/glossary.html#term-hashable
* https://www.python-course.eu/python3_sets_frozensets.php


```python
# CREATING SETS
A = set()    # empty set
B = {4, 5}   # non-empty set

L = [3,5,7,9]
C = set(L)   # non-empty set with iterable constructor; note the difference in member order

A.add(8) # add a member to a set (i.e. sets are mutable)
A.add(5)

C.remove(5)  # remove a member from a set

print(A, type(A), len(A), 5 in A)
print(B, type(B), len(B), 5 in B)
print(C, type(C), len(C), 5 in C)
```

```python
# CREATING FROZEN SETS
A = set()                    # empty set
B = frozenset([1, 2, 3])     # non-empty FROZEN set using iterable constructor
C = frozenset([4, 5, 6])     # ditto
D = set([7, 8, 9])           # non-empty REGULAR set using iterable constructor

# frozen sets can be members of a set
A.add(B)
A.add(C)

# removing these comments will cause an error - so play around with it to see what it does :)
# A.add(D) <=== however, this causes an error because a member of a set must be hashable
# B.add(4) < === this causes an error because a frozen set is immutable (you cannot change it after it is created)

print(A)
```

### A.1 Mathematical Set Operations
In Python, sets can be used to carry out mathematical set operations like union, intersection, difference and symmetric difference. We can do this with operators or methods.

* https://www.programiz.com/python-programming/set


```python
# UNION
A = {3, 4, 5}
B = {4, 5, 6}
C = A | B
print(C)     # <=== {3, 4, 5, 6}

C = A.union(B)
print(C)     # same result

# INTERSECTION
C = A & B
print(C)     # <=== {4, 5}

C = A.intersection(B)
print(C)     # same result


# DIFFERENCE
C = A - B
print(C) # <=== {3} i.e. what is in A that is not in B

C = A.difference(B)
print(C)

C = B - A
print(C) # <=== {6} i.e. what is in B that is not in A

C = B.difference(A)
print(C)

# SYMMETRIC DIFFERENCE
C = A ^ B
print(C) # <=== {3, 6}

C = A.symmetric_difference(B)
print(C) # same result


# DISJOINT SETS

result = A.isdisjoint(B)
print(result)             # <=== False because there are common elements in A and B

A = set([1, 2, 3])
B = set([4, 5, 6])
print(A.isdisjoint(B))    # <=== True because there are NO common elements in A and B
```
