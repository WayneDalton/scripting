# Working with Data using Python Data Types
In this unit the focus will be on the data types available as part of the core language.  These are considered as standard types and are built into the interpreter. https://docs.python.org/3/library/stdtypes.html

1. Numeric Types - There are three distinct numeric types: integers, floating point numbers, and complex numbers. In addition, Booleans are a subtype of integers. Integers have unlimited precision. Floating point precision and internal representation for the machine on which your program is running is available in sys.float_info. Complex numbers have a real and imaginary part, which are each a floating point number.


2. Sequence Types - Sequences are finite, ordered collections of objects indexed by non-negative numbers.  They are distinguished by their mutability (i.e. their ability to be changed after they are created).  Some cannot change (aka immutable sequences), such as strings, bytes, tuples and ranges.  Some can change (aka mutable sequences), such as lists and byte arrays.


3. Set Types - Sets are unordered, finite collections of unique, immutable objects. As such, they cannot be indexed by any subscript. Common uses of sets include membership testing, removing duplicates from a sequence, and computing mathematical operations such as intersection, union, difference, and symmetric difference. Sets are mutable, but frozen sets are not.


4. Mapping Types - A mapping object maps hashable values to arbitrary objects. Mappings are mutable objects. There is currently only one standard mapping type, the dictionary. Dictionaries are finite sets of key-value pairs of objects indexed by nearly arbitrary values (i.e. only keys that are hashable).

In this unit we will explore these built-in Python 3 data types by tackling a variety of problems.

## A. Dictionary Problems
A mapping object maps $hashable$ values (keys) to arbitrary objects (values). Mappings are mutable objects. There is currently only one standard mapping type, the dictionary. A mapping is a container object that supports arbitrary key lookups, much like we would use a glossary of terms, the table of contents for a book, or an actual dictionary or thesaurus to look up the meaning or context of a word.

An object is hashable if it has a hash value which never changes during its lifetime and can be compared to other objects (they compare equal if and only if they have the same hash value). Hashability makes an object usable as a dictionary key and a set member, because these data structures use the hash value internally.

### A.1 South African Official Languages
In unit 5 we noted that South African has 11 official languages.
* https://en.wikipedia.org/wiki/Languages_of_South_Africa

Now, instead of using two separate lists (e.g. languages and percents), we can use a single dictionary to model this data, as depicted below:



```python
# MODELING THE OFFICIAL LANGUAGES OF SOUTH AFRICA

languages = {}  # create an empty dictionary (you can also use dict() constructor function)

languages["Zulu"] = 22.7
languages["Xhosa"] = 16.0
languages["Afrikaans"] = 13.5
# for brevity we ignored some data - feel free to include them in your own code
languages["Ndebele"] = 2.1

print(languages)
```

```python
# if you want to create a dictionary with prepopulated values
languages = {"Zulu":22.7, "Xhosa":16.0, "Afrikaans":13.5, "Ndebele":2.1}
print(languages)

percent_zulu_speakers = languages["Zulu"]
print("% people who speak Zulu =", percent_zulu_speakers)

percent_afri_speakers = languages["Afrikaans"]
print("% people who speak Afrikaans =", percent_afri_speakers)

# or you can use a variable to select the language
lang = "Xhosa" # <=== change this to see how this affects the output
percent_lang_speakers = languages[lang]
print("% people who speak", lang, "=", percent_lang_speakers)
```

```python
# HOW DO I FORMAT A FLOATING POINT TO LOOK LIKE A CURRENCY?
amount = 150
rate = 12.598123
new_amount = amount * rate

print(new_amount)                          # raw floating point
print("R{:.2f}".format(new_amount))        # closer, but not quite there
print("R{:10,.2f}".format(new_amount))     # accounting format; 10 spaces, include comma, 2 decimal places for float
```

### A.2 Just Zip It
Some times you have data in separate lists, and you want them to be combined into a dictionary.  Python provides a builtin function zip() to help achieve this.  Official documentation and online explanations of the zip function can be a bit daunting to someone who is new to Python, so for now lets think about zip() as a way to create an object which is the result of associating (or mapping) one sequence to another.  Lets take two lists for example:

* List 1 = ["holy", "good", "face", "gat"]
* LIst 2 = ["smoke", "grief", "plant", "vol"]

These words are unrelated, but we can produce someting new when we associate these words with each other.  The zip() function creates a new object 

For further reading:
1. https://docs.python.org/3/library/functions.html#zip


```python
list1 = ["holy", "good", "face", "gat"]
list2 = ["smoke", "grief", "plant", "vol"]

z = zip(list1, list2) # z is now a zip object
print(z) # by itself it is less useful, but we can do things with it that are useful

list3 = list(zip(list1, list2))
print(list3) # a list of tuples where each word is associated

dict3 = dict(zip(list1, list2))
print(dict3) # a dictionary where each word is mapped to an index

# here is a glimpse of how this dictionary could be used
for key in dict3.keys():
    print(key, dict3[key])

```

```python
languages = [
    "Zulu", "Xhosa", "Afrikaans", "English", "Northern Sotho", "Tswana", 
    "Sesotho", "Tsonga", "Swati", "Venda", "Ndebele", "Sign", "Other"
]

# count of 1st langauge speakers for each subject
totals = [11587374, 8154258, 6855082, 4892623, 4618576, 4067248, 3849563, 2277148, 1297046, 1209388, 1090223, 234655, 828258]

# map language to the total for that language e.g. Zulu => 11,587,374
d_lang = dict(zip(languages, totals))

# first we get the total population by taking the sum of all the values in the dictionary
total_population = sum(d_lang.values())
print("Total Population:", total_population)

lang = "English" # <== change this to any other language from the list above (i.e. languages)
total_lang_speaker = d_lang[lang]
percent_lang_speaker = (total_lang_speaker / total_population) * 100
print(lang, total_lang_speaker, percent_lang_speaker)
```
