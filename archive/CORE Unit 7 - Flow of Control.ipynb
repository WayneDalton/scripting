## Blocks of Code are Indented

Source: https://www.python-course.eu/python3_blocks.php

A block is a group of statements in a program or script. Usually it consists of at least one statement and of declarations for the block, depending on the programming or scripting language. A language, which allows grouping with blocks, is called a block structured language. Generally, blocks can contain blocks as well, so we get a nested block structure. A block in a script or program functions as a mean to group statements to be treated as if they were one statement. In many cases, it also serves as a way to limit the lexical scope of variables and functions.

Python programs get structured through indentation, i.e. code blocks are defined by their indentation. This is a language requirement not a matter of style, but observing this principles makes it easier to read and understand other people's Python code; which is a very good thing.



All statements with the same distance to the right belong to the same block of code, i.e. the statements within a block line up vertically. The block ends at a line less indented or the end of the file. If a block has to be more deeply nested, it is simply indented further to the right. 

NOTE: most text editors will have some limited form of autoindent; some use tabs and others use a fixed number of spaces.  For example, when you defined a function.  It is always your responsibility as programmer to ensure indentation is correctly applied.

## Conditional Execution
In Unit 3 the Boolean data type was introduced.  Booleans are either True or False.  Conditional execution exploits this fact by evaluation an expression and then executing code based on whether the expression was True or False.

### The Simple IF statement
The simplest form of conditional execution is a binary if-statment:

```python
if x < 0:
    print('a negative number')
```

### The Chained IF statement
For more complex situations, you need to have several IF-statements.  In Python you can chain several IF-statements together (by using the "else if" or elif statement) and also include an optional ELSE clause.  If none of the chained if-statements evaluate to True then the "else" code block is executed (or the code does nothing if the else-block is omitted).  This is illustrated below:

```python
if x < 0:
    print('a negative number')
elif x > 0:
    print('a positive number')
elif x == 0:
    print('a zero')
else:
    print('not sure what that was!')
```

### Truth Expressions
Boolean algebra, which is a branch of mathematics, provides us with a form of reasoning where the values of the variables are either True or False.  This is helpful when programming conditional execution, and also to make our code more readable and understandable to others.

In the code below, several Boolean variables (e.g. hasSugar) have be configured to model the reality of a given scenario (i.e. drinking coffee - a fine enterprise even if I say so myself).  In addition, several truth expressions have been put in place (e.g. isPerfect) which are reduced to a Boolean value of either True or False.

When we invoke these truth expressions in code, it make the code much more readable to other humans.  You should see this in the Coffee Evaluation Process section of the code below.

```python
# Making a drinkable cup of coffee

# COFFEE CONFIGURATION PANEL
hasSugar = False
hasMilk = True
hasRicoffy = False
hasNescafe = False
hasDouweEg = True
points = 0

# COFFEE APPRECIATION RULES AKA TRUTH EXPRESSIONS
isPerfect = hasMilk and not hasSugar and hasDouweEg
isBad = hasSugar and hasRicoffy
isOkay = hasMilk and not hasSugar and (hasNescafe or hasDouweEg)

# COFFEE EVALUATION PROCESS
if isPerfect:
    print('This coffee is perfect!')
    points += 100
elif isOkay:
    print('This coffee is good')
    points += 50
elif isBad:
    print('This coffee is BAD :[')
    points -= 50
else:
    print('Insufficient data to form an opinion')
    points += 0

print('Score:', points)
```

## Repetitive Execution aka Iteration
Most tasks, whether they are simple or complex, involve some form of repetition.  Computers are good at performing repetitive tasks error-free, but humans sadly are not.

In computer programming jargon, when you walk through a sequence of items, you "iterate across" these items.  An object that can be used in iteration is called an iterable.  Objects that are of a collection type in Python can be used as a iterator e.g. string, list, tuple, dict, set, and frozenset types.


### Definite Iteration
If you need to repeat something for a fixed number of time e.g do this 5 times, then we need definite iteration (i.e. repetition that will definitely end at a specific point).  We can rely on this to only happen 5 times.  Python provides the FOR loop for definite iteration.

Python supports definite iteration over iterables and iterators:

* __Iterable__ - data types that are collection or container types such as string, list, tuple, dict, set, and frozenset types). Many objects that are built into Python or defined in modules are designed to be iterable. Even user-defined objects can be designed in such a way that they can be iterated over, as you will discover in the OOP programming unit.

* __Iterators__ - essentially a value producer that yields successive values from its associated iterable object. The built-in function next() is used to obtain the next value from in iterator.  
    * An iterator retains its state internally and knows which values have been obtained already, so when you call next(), it knows what value to return next. 
    * If all the values from an iterator have been returned already, a subsequent next() call raises a StopIteration exception. Any further attempts to obtain values from the iterator will fail.  
    * You can only obtain values from an iterator in one direction. You can’t go backward. There is no prev() function.

#### Reading
* https://realpython.com/python-for-loop/


```python
""" COLLECTIONS-BASED ITERATION """

# definite iteration over a list
letters = ['a', 'b', 'c', 'd', 'e']
for letter in letters: # <= iterate over letters
    print(letter)

# definite iteration over a tuple - this iterates for a fixed number of times
numbers = (1, 2, 3, 4, 5)
for number in numbers: # <= iterate over numbers
    print(number)
```

```python
""" ITERATOR-BASED ITERATIONS """
# letters = ['a', 'b', 'c', 'd', 'e']
# numbers = (1, 2, 3, 4, 5)

# We create iterators from the list and tuple from the previous example
iLetter = iter(letters)
iNumber = iter(numbers)

# you pass iterators to the next() function
print('letter:', next(iLetter))
print('number:', next(iNumber))

print('letter:', next(iLetter))
print('number:', next(iNumber))

# you can access an entire iterator by converting it's internal state to a list
print('remaining letters', list(iLetter))
print('remaining numbers', list(iNumber))
```

### Indefinite Iteration
Sometimes, we need to repeat something and continue to do so until a condition is met.  We may not even be sure of when or if this condition will ever be met.  We cannot reliably estimate how many iterations will take place, if any, but we can rely on the fact that if the condition is met then execution will terminate (if it started in the first place).  Python provides the WHILE loop for indefinite iteration.

#### The Stop Case
If you turn on a garden sprinkler, it will continue to water your garden until you turn it off or you run out of water (at which point it fails to water your garden any longer).  Most of us dont really want to stand and water the garden by hand, so we would rather deploy an automated sprinkler system which will water the garden whilst we are asleep (or watching rugby).  In this case, you need to give the automation system set times to switch on and to switch off; especially if you dont want to be fined for wasting water.

When using indefinite iteration, we must ensure there is a natural way for the iteration to terminate.  This is called a __stop case__ and is most commonly expressed as a _boolean_ expression of some form.

#### Reading
* https://realpython.com/python-while-loop/

```python
x = 0
while True:
    x += 1
    print(x)
    if x > 10: break
print('done')
```

### Rules of Thumb

Reading: https://treyhunner.com/2018/02/python-range-is-not-an-iterator/

* If you know you can loop over something, it’s an **iterable**.

* If you know you can pass something to the next function, it’s an **iterator**.

* If you know the thing you’re looping over happens to compute things as you loop over it, it’s a **_lazy_ iterable**.

* Python 3’s range object is not an iterator.
