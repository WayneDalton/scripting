## B. Simple Mathematical Formulae
A function is an expression, rule, or law that defines a relationship between one variable (the independent variable) and another variable (the dependent variable). Functions are ubiquitous in mathematics and are essential for formulating physical relationships in the sciences. The modern definition of function was first given in 1837 by the German mathematician Peter Dirichlet:

    If a variable y is so related to a variable x that whenever a numerical value is assigned to x, there is a rule according to which a unique value of y is determined, then y is said to be a function of the independent variable x.

This relationship is commonly symbolized as y = f(x). In addition to f(x), other abbreviated symbols such as g(x) and P(x) are often used to represent functions of the independent variable x, especially when the nature of the function is unknown or unspecified. https://www.britannica.com/science/function-mathematics


```python
def f(x):
    """ y is equal to twice the size of the independent variable x """
    return 2 * x

x = 4
y = f(x)
print(x, y)
```

## C. In an Argument, Order Matters
Typically, functions take positional arguments.  The order matters because normally it affects the rule contained within in the function itself.  In the function below, the result is determined by the order of the arguments.

```python
def f(x1, x2):
    """ subtract x2 from x1 """
    print('{} - {}'.format(x1, x2))
    return x1 - x2

y = f(4, 8)
print(y)

y = f(8, 4)
print(y)
```

Sometimes, one of the arguments of a function might be a persistent value.  For example, VAT rate is needed each time you calculate the sales price of an item but the rate remains the same (until it is changed under special circumstances).  Recently VAT in South Africa increased by 1%.  In some circumstances the VAT you charge changes, as is the case with zero-rated items.

```python
def price(amount, rate):
    """ 
        return sales price based on cost amount and VAT rate
        arg1 amount: product cost (float)
        arg2 rate: VAT percentage (float) e.g. 0.15 is 15%
    """
    return amount * (1+rate)

p = price(4000.0, 0.15)
print("R {:,.2f}".format(p))

p = price(125.56, 0.0)
print("R {:,.2f}".format(p))

```

## D. Functional Programming Concepts

Python 3 has lambda expressions (nameless functions) which implements lambda calculus.

Lambda calculus was introduced in the 1930s by Alonzo Church (who supervised the PhD of Alan Turing amongst other eminent students). In the 1960s, Peter Landin was the first make a connection between Lambda calculus and programming languages; enabling the development of functional programming languages (e.g. Haskell) and to a diverse range of programming applications and concepts. https://en.wikipedia.org/wiki/Peter_Landin

Computerphile's video on Lambda Calculus: https://www.youtube.com/watch?v=eis11j_iGMs (13 minutes)

Whilst lambda calculus is very important for computer science, the inclusion of lambda expressions in Python (and the associated map(), filter() and reduce() functions) is  often debated and actively avoided by a growing number of programmers because they are not strictly necessary and are subject to overuse and to misuse.


### Should you ever use lambda expressions?

According to Trey Hunner, the use of lambda expressions problematic because:

* they require a working knowledge of lambda calculus or functional programming in languages like Lisp
* they impair the readability of Python code (no function names or documentation, dont support language tuple unpacking))
* they can often be replaced with already existing functions in the standard libray or built-in to Python

On the other hand, a def statement is often more understandable, Python also has a number of features that can be used to replace lambda expressions, including special syntaxes (comprehensions), built-in functions (sum), and standard library functions (in the operators module).

If you must use lambda expressions, ensure your situation meets all four of these criteria:

* The operation you’re doing is trivial: the function doesn’t deserve a name
* Having a lambda expression makes your code more understandable than the function names you can think of
* You’re pretty sure there’s not already a function that does what you’re looking for
* Everyone on your team understands lambda expressions and you’ve all agreed to use them

If any of those four statements don’t fit your situation, I’d recommend writing a new function using def and (whenever possible) embracing a function that already exists within Python that already does what you’re looking for.

* Read https://treyhunner.com/2018/09/stop-writing-lambda-expressions/
* For a more thorough argument, read http://www.paulbrownmagic.com/blog/vslambda

```python
numbers = [2, 3, 4, 5]

def square(x):
    return x * x

L = []
for number in numbers:
    L.append(square(number))
print(L) # [4, 9, 16, 25]

# using map(aFunction, aSequence) which applies the function across every element of the list
L = list(map(square, numbers))
print(L) # same output

# now with lambda function in the map(aFunction, aSequence) call
L = list(map(lambda x : x*x, numbers))
print(L) # same output

# now with list comprehension (i.e. without the need of either of these methods)
L = [x*x for x in numbers]
print(L) # same output
```

```python
A = [1,3,5,6,7,8,9]
B = [6,8,9,10,12]

def intersect(a, b):
    result = []
    for item in a:
        if item in b:
            result.append(item)
    return result

print(intersect(A, B)) # [6, 8, 9]

# which items in A and B are the same using filter(aFunction, aSequence)
L = list(filter(lambda x : x in A, B))
print(L) # same result

# now with list comprehension
L = [x for x in A if x in B]
print(L) # same result
```
