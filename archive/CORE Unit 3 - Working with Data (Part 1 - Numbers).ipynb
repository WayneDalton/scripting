# Working with Data using Python Data Types
In this unit the focus will be on the data types available as part of the core language.  These are considered as standard types and are built into the interpreter. https://docs.python.org/3/library/stdtypes.html

1. Numeric Types - There are three distinct numeric types: integers, floating point numbers, and complex numbers. In addition, Booleans are a subtype of integers. Integers have unlimited precision. Floating point precision and internal representation for the machine on which your program is running is available in sys.float_info. 2. Sequence Types - Sequences are finite, ordered collections of objects indexed by non-negative numbers. Some can change (aka mutable sequences), such as lists and byte arrays.


3. Set Types - Sets are unordered, finite collections of unique, immutable objects. As such, they cannot be indexed by any subscript. Common uses of sets include membership testing, removing duplicates from a sequence, and computing mathematical operations such as intersection, union, difference, and symmetric difference. Sets are mutable, but frozen sets are not.


4. Mapping Types - A mapping object maps hashable values to arbitrary objects. Mappings are mutable objects. There is currently only one standard mapping type, the dictionary. Dictionaries are finite sets of key-value pairs of objects indexed by nearly arbitrary values (i.e. only keys that are hashable).

### A.1 The Whole Number and Nothing But the Number
An integer is a whole number that can be written without a fractional component.  An integer can be positive, negative or zero.  Integers can be precisely represented by computers using binary numbers.




#### Unsigned Integers
Computers store integers internally as binary numbers, even though we often display them as decimal numbers.  Binary means "base 2" in the same way as decimal means "base 10".  So the value "100" in decimal represents the value one hundred, the value "100" in binary represents the value four.

A binary number consists of a fixed number of "bits" (binary digits).  A byte consists of 8 bits (half a byte is called a "nibble" which is 4 bits long).  Contemporary computer architectures typically accomodate 8-bit, 16-bit, 32-bit and 64-bit numbers.  These are normally referred to as byte, word, double word and quad words.

A binary number can store an integer value precisely, as long as you use sufficient bits in that representation.  


An 8-bit binary value is depicted above.  Each bit occupies a position in the byte, identified by its index; ranging from the least significant bit (index=0) to the most significant bit (index=7).  Because we are working with base 2, the legal range for each digit is {0, 1}. You can therefore only write a 1 or a 0 in each bit location.  Note the following implication when adding numbers:
1. 0 + 1 = 1 (no carry)
2. 1 + 1 = 0 (carry 1)

This is similar to base 10, which we are more familiar with.  Legal range of digits is ${0, 1, ..., 9}$. If an addition operation yields a result that is greater than 9, then it carries 1 and leaves the remainder (which is <= 9) behind.

For this reason, numerical value progress (using n bits) follows a consistent pattern. For example, n=3:
* 0 = 000
* 1 = 001
* 2 = 010
* 3 = 011
* 4 = 100
* 5 = 101
* 6 = 110
* 7 = 111

A binary bit pattern such as "01001010" can be converted into a decimal integer by:
1. Observing the bit position of all of the "1"s; in the example above that would be 1, 3 and 6
2. A "1" in bit position 1 represents the value of "two the the power of 1" or the value 2; similarly a "1" in position 3 represents the value of "two to the power of 3" or the value 8; etc.
3. Adding up the indvidual values (viz. 2 + 8 + 64) produces the value represented by the entire bit pattern


```python
# python will convert integers into their equivalent binary format for you
x = 75
y = bin(x)

# you can print a number in it binary format
print('{0} in binary format is {0:b}'.format(x))

# you can explicitly convert it to binary and then print it
print('{} in binary format is {}'.format(x, y))

# you can supply a value in it's binary format
x = 0b1001010
print(x)
```

#### Signed Integers
The definition of an integer stated that it included negative values as well.  So how does one represent a negative number in binary?  The method of representing a negative number in binary is called the process of "two's complement".  Assume we want to represent -74 as an integer.

Two's complement works as follows:
1. Take the postive value and flip the bits - in this case we already know that "74" is "01001010"; flipping the bits results in "10110101"; i.e. if the bit is "1" change it to a "0" and vice versa.
2. Add 1 - adding one to this result produces "10110110" which is -74 in binary format (do you understand why?)



Note that when representing signed integers (i.e. integers that hold positive and negative values), the most significant bit (bit position 7) will ALWAYS be 1.  The astute reader will also note that unsigned integers can represent a larger range of values (e.g. 0...255) but signed integers cannot due to the sign (e.g. -128,...,-1,0,1,...,127)

### The Whole Truth and Nothing but the Truth
In the code above, we encounter truth values for the first time; viz. True and False.  These are objects of type Boolean.  They are implemented as integers; i.e. False is implemented as integer zero, and True is implemented a non-zero integer.

In general, any object in Python can be tested for a truth value.
1. If two values x and y are tested, and x is greater than y, then x > y evaluates to True; otherwise it evaluates to False
2. If two values x and y are tested, and x is smaller than y, then x < y evaluates to True; otherwise it evaluates to False
3. If two values x and y are tested, and x is equal to y, then x == y evaluates to True
4. Similarly, x and y can be evaluated using <= (less than or equal) and >= (greather than or equal), yielding True is those conditions hold otherwise False.



```python
# truth values True and False are of type Boolean
print(type(True), type(False))

# True is represented as 1; False is represented as 0
print(int(True), int(False))

# various expressions that evaluate to False have zero value or zero length
print(bool(0), bool(None), bool(False), bool(""))

# various expressions that evaluate to True; everything that is not False is True
print(bool(-1), bool(1), bool(True), bool("xxx"))

x = 2
y = 4
z = 6

print(x <= 4) # true; x is less than 4
print(y <= 4) # true; y is equal to 4
print(z <= 4) # false; z is larger than 4
```

### A.2 Let's Get Real and Rational about Fractions
Integers do not represent all the numbers we are interested in.  There is a much larger collection of values that require a fractional component which cannot be represented by integers; these are floating point numbers or "floats" as they are known in many programming languages.  

Floating point numbers are the numbers that lie between the integers.  If we only use a single decimal place then between 0 and 1 there are 9 other numbers; viz. {0.1, 0.2, ... , 0.9}.  Of course, this set of numbers grows rapidly as you increase then number of decimal places in your precision (i.e. how many bits you use to represent a floating point number - e.g. single precision uses 32 bits and double precision uses 64 bits).



Unlike integers, computers cannot represent floating point numbers using binary numbers; well, not precisely. So computers use a clever method to approximate floating point values using binary numbers.  This is a language-independent issue.  

To understand this problem in more detail, we must first realise that humans think and do math using decimal numbers whilst computers do the same using binary numbers.  Here is an example to illustrate this:

1. Decimal (base 10): decimal places are in powers of 10; this is how we think about fractions: $ 0.125 = {1\over10} + {2\over100} + {5\over1000} $


2. Binary (base2): decimal places are in powers of 2; this is how computers think about fractions: $ 0.125 = {0\over2} + {0\over4} + {1\over8}$

Unfortunately, not all decimal fractions have identical binary equivalents.  This can lead to some pretty bizarre results under certain circumstances. Here is an example where the approximation leads to a bizarre result:


```python
x = 0.1 + 0.1 + 0.1
y = 0.3
print(x==y) # decimal thinkers would believe this is true, but Python reports this as false
```

The question is why.

Python uses IEEE-754 double-precision floating point numbers.  The range and overflow handling is therefore dependent on the underlying machine and interpreter architecture.  



As you can see, from the picture above, the internal representation of "0.1" is precise to only a few decimal places.  This means that there is a small error between the value entered and the value stored, because it is being represented internally as a binary number.



This is why comparisons like this (see python code above) will give bizarre result since internally these numbers are approximations.  Understanding the IEEE-754 specification is beyond the scope of this module, but you can use link 2 below to play around with other numbers for deeper insight into the impact this will have.

The goal here is not to fully understand IEEE-754, but rather to:

#### The Complex Conjugate - A Mirror Image
The conjugate of the complex number $z=x+yi$ is given by $z=x−yi$.  Effectively, it is the "mirror image" of z, and denoted by $\overline{z}$.  The product of a complex number $z=x+yi$ and its conjugate is always a positive real number and equals the square of the magnitude of each.

$$z.\overline{z}=x^2+y^2$$

Note: The division of two complex numbers can be accomplished by multiplying the numerator and denominator by the complex conjugate of the denominator.



```python
# construct a complex number z = 2-3j
a = 2
b = -3
z = complex(a, b)

# calculate the conjugate "mirror image" of z
zc = z.conjugate()

# z multiplied by its conjugate yields a^2 + b^2; x therefore will always have 0j
x = z * zc
print(z, zc, x)
```

```python
z1 = complex(3,2)
z2 = complex(4,-3)

print(z1/z2) # <=== easy division which gives you correct answer

# if you did this by hand - this is the method you would follow
above_the_line = z1 * z2.conjugate() # multiply numerator by conjugate of denominator
below_the_line = z2 * z2.conjugate() # multiply denominator by its own conjugate (which is always a real number)

print('above', above_the_line)
print('below', below_the_line)

r = above_the_line.real / below_the_line.real
i = above_the_line.imag / below_the_line.real

# same result as above but using the long-hand method
z3 = complex(r, i)
print(z3)
```

#### Getting to the Root of the Problem
The history of complex numbers is tightly coupled to the pursuit of solving equations of the third degree; commonly referred to as cubic equations; in particular, finding the roots of a cubic equation.  For more detail refer to these online resources:
1. http://www.math.uri.edu/~merino/spring06/mth562/ShortHistoryComplexNumbers2006.pdf
2. https://www2.clarku.edu/faculty/djoyce/complex/cubic.html

For example: $f(x)=x^3-4x^2+x+26$



It is fairly easy to see from a plot of this equation that it has at least one real root of $x = -2$, but it turns out that it also has a complex root. To find all the roots, we rewrite the equation with the factor $(x+2)$, and focus on the second factor (which is a quadratic equation) i.e. $(x+2)(x^2-6x+13)=0$

To solve for x in the quadratic equation $(x^2-6x+13)$ we use the quadratic formula $x={-b\pm\sqrt{b^2-4ac}\over{2a}}$ and substitute the terms ${a=1, b=-6, c=13}$:

$x={6\pm\sqrt{(-6)^2-(4\times(1)\times13)}\over{2\times(1)}} ={6\pm\sqrt{-16}\over{2}} ={6\pm\sqrt{16}\sqrt{-1}\over{2}} ={6\pm4i\over{2}} =3\pm2i$.

Therefore the roots of this cubic equation are real and complex; viz. $x=-2$ and $x=3\pm2i$.

#### Polar Coordinates
A common operation involving complex numbers, is to convert from Cartesian to Polar coordinates.  In the Cartesian Plane, the complex number $z=x+yi$ is plotted using the distances of $x$ and $y$ in 2D space.  Polar coordinates specify the same point, but uses a distance $r$ and an angle $\theta$ relative to the origin.



Pythagorus' theorem provides the following relationships:
1. $r = \sqrt{x^2 + y^2}$


2. $\theta = tan^{-1}{y\over{x}}$

Python makes the conversion between Cartesian and Polar coordinates really convenient and easy.  The cmath module is always available and provides access to mathematical functions for complex numbers. A Python complex number z is stored internally using rectangular or Cartesian coordinates.  
* https://docs.python.org/3/library/cmath.html

```python
from math import degrees
from cmath import polar

x = 3
y = 4

z = complex(x, y)
r, rad = polar(z)
deg = degrees(rad)

print('{} in polar form is r={:.1f} angle={:.1f}'.format(z, r, deg))
```
