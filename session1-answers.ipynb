There are 52 weeks a year. Abel earns R25.95 per hour and works a 40-hour week. Calculate his annual income. R53,976

```python
name = "Abel"
weeks = 52 # in 1 year
wage = 25.95 # per hour
hours = 40 # per week

# annual income is 52 weeks @ R25.95 per hour for 40 hours
income_per_week = wage * hours
annual_income = income_per_week * weeks

# formatted output
print(f"{name} works for {hours * weeks:,d} weeks @ R{wage:.2f} per week = R{annual_income:,.2f}")
```

VAT is 15%. After shopping at Spar, your till slip total is R1750.44 including VAT. What was the VAT amount you paid? R228.32 (R1,522.12 excl) 
hint: https://calculatevat.co.za/index.php

```python
total_incl = 1750.44 # till slip
vat = 0.15 # 15%

total_excl = total_incl / (1+vat)
vat_paid = total_incl - total_excl

print(f"Total (inclusive): R {total_incl:7.2f}")
print(f"Total (exclusive): R {total_excl:7.2f}")
print(f"Total VAT paid is: R {vat_paid:7.2f}")
```

Given a right-angled triangle, with perpendicular length of 6 and a base length of 8. Calculate the length of the hypothenuse using Pythagoras’ Theorem. 10.0 
hint: https://byjus.com/maths/pythagor

Write a program that prompts the user for a temperature in Celsius and then converts the temperature to Fahrenheit, and prints out the converted temperature. hint: https://www.checkyourmath.com/convert/temperature/celsius_fahrenheit.php

```python
temp_c = float(input("Enter a temperature (in Celsius): "))
temp_f = (temp_c * 1.8) + 32
print(f"{temp_c:.2f} celcius is equivalent to {temp_f:.2f} fahrenheit")
```
