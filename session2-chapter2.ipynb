# SEQUENCE: Straight-line Code
One statement is executed after the other in the order they are written. In other words, the program executes in a straight line.  This is  good but not very powerful.


# SELECTION: Conditional Execution
A conditional process step is executed on when the execution condition is met.


# ITERATION: Repetitive Execution
A program loops until a condition is met.
* Indefinite repetition - WHILE Loop (keep looping until condition is met)
* Definite repetition - FOR Loop (keep looping for a fixed, determined number of iterations)
